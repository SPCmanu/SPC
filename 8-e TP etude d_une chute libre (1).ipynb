**Document 2 : La chute libre**

Par définition, on dit qu’un corps est en chute libre si la seule force à laquelle il est soumis est son poids. Par exemple, les astronautes à bord de l’ISS sont en chute libre. Une bille lâchée sans vitesse initiale peut être considérée comme en chute libre durant les trois premiers mètres de sa chute.

**Document 4 : Le principe d’inertie**

L’expression mathématique du principe d’inertie est : 
$\vec{F_{extérieurs}}= \vec{0}$
avec $\vec{F_{extérieurs}}$ qui représente la somme des forces s’exerçant sur le système.

Remarque : $\vec{v} = \vec{0}$ (cas d’un système au repos) est un cas particulier de $\vec{v}$ = constant (cas d’un mouvement rectiligne uniforme).

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

# Création de la fonction permettant de tracer un vecteur vitesse
def vecteur_vitesse(x, y, dt, i):
    vx = (x[i+1] - x[i]) / (dt)
    vy = (y[i+1] - y[i]) / (dt)
    plt.quiver(x[i], y[i], vx, vy, angles="xy", scale_units="xy", scale=25, color="green")

# Pointage des positions d'un objet en chute libre toutes les 0,1 s.
dt = 0.1
x = np.array([0, 0, 0, 0, 0, 0, 0, 0])
y = np.array([3.00, 2.95, 2.80, 2.56, 2.22, 1.77, 1.23, 0.60])

# Affichage
plt.figure(figsize=(12,10))
plt.plot(x, y, "b+")
for i in range (0, 7):
    vecteur_vitesse(x, y, dt, i) # Tracé du vecteur vitesse au point i

plt.xlabel("Position x")
plt.ylabel("Position y")
plt.title("Chute libre sans vitesse initiale")
plt.show()

v=[]
for i in range (len(y)-1):
    vi=-(y[i+1]-y[i])/(dt)
    v.append(vi)

t=[]
for n in range (len(x)-1):
    tn=(n*dt)
    t.append(tn)

v=np.array(v)
t=np.array(t)
    
print(v)
print(t)

#regression linéaire
from scipy import stats
slope, intercept, r_value, p_value, std_error = stats.linregress(t, v)
print ('slope {0:.2f}'.format(slope))
print('intercept {0:.2f}'.format(intercept))
vmodel = slope*t+intercept
print('v= {0:.2f}'.format(slope),'x t')
print ('Le coefficient de corrélation r vaut {0:.4f}'.format(r_value))


plt.show()

# Affichage
plt.figure(figsize=(12,10))
plt.plot(t,v,'r+',label='v=f(t)')
plt.legend()
plt.xlabel("temps t")
plt.ylabel("vitesse v")
plt.title("Chute libre sans vitesse initiale")
plt.grid()

#model
plt.plot(t,v,'r+',label='v=f(t)')
plt.plot(t,vmodel,'b',label='modèle linéaire')
plt.legend()
plt.xlabel("intensité t (s)")
plt.ylabel("tension v (m/s)")
plt.grid()

plt.show()
```
